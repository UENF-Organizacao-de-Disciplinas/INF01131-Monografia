```python
import openpyxl
from openpyxl.styles import PatternFill
import re

# Load the workbook
workbook = openpyxl.load_workbook('cores.xlsx')


# Get the active sheet
sheet = workbook.active

# Iterate through each cell in the sheet
for row in sheet.iter_rows():
  for cell in row:
    # Check if the cell value is a hex color code
    if isinstance(cell.value, str) and cell.value.startswith('#'):
      # Check if the cell value is a valid aRGB hex value
      if re.match(r'^#([A-Fa-f0-9]{6}|[A-Fa-f0-9]{8})$', cell.value):
        # Create a pattern fill with the hex color code
        fill = PatternFill(start_color=cell.value.lstrip('#'), end_color=cell.value.lstrip('#'), fill_type='solid')
        # Apply the fill to the cell
        cell.fill = fill
        print(f"Valid color code: {cell.value}")
      # else:
        # print(f"Invalid color code: {cell.value}")
    # else:
      # print(f"Invalid color code: {cell.value}")
workbook.save('cores.xlsx')
```

```python
import openpyxl
from openpyxl.styles import PatternFill
import re

# Load the workbook
workbook = openpyxl.load_workbook('cores.xlsx')

# Select the active sheet
sheet = workbook.active

# Iterate through each cell in the sheet
for row in sheet.iter_rows():
  for cell in row:
    # Check if the cell value is a hex color code
    if isinstance(cell.value, str) and cell.value.startswith('#'):
      # Check if the cell value is a valid aRGB hex value
      if re.match(r'^#([A-Fa-f0-9]{6}|[A-Fa-f0-9]{8})$', cell.value):
        # Create a pattern fill with the hex color code
        fill = PatternFill(start_color=cell.value.lstrip('#'), end_color=cell.value.lstrip('#'), fill_type='solid')
        # Apply the fill to the cell
        cell.fill = fill
      else:
        print(f"Código de cor inválido: {cell.value}")
    else:
      print(f"Código de cor inválido: {cell.value}")
workbook.save('cores.xlsx')
```
